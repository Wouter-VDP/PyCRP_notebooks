```python
from os.path import join, exists
import joblib
import numpy as np
import matplotlib.pyplot as plt
```

```python
Background,edges0 = joblib.load("Background_workingpoint_August.joblib") # Rate 1Hz
Ba133,edges1 = joblib.load('Ba133.joblib')
Co60,edges2 = joblib.load('Co60_August.joblib') # Rate 2.65Hz
AmBe, edges3 = joblib.load("AmBe_August.joblib") # Rate 2Hz

assert(np.all(edges1==edges0))
assert(np.all(edges2==edges1))
assert(np.all(edges3==edges2))

mids = (edges0[1:]+edges0[:-1])/2

rate_background = 0.97
rate_AmBe = 1.97
rate_Co60 = 2.65

background_normalised = Background/sum(Background)*rate_background
```

```python
fig, ax  = plt.subplots(figsize=(12,4), constrained_layout=True)

#ax.plot(mids,Ba133,label="Ba133",alpha=0.5)
#ax.plot(mids,Co60,label="Co60",alpha=0.5)
ax.plot(mids,AmBe,label="AmBe",alpha=0.5)
ax.plot(mids,Background,label="Background",alpha=0.5)
ax.set_yscale("log")
ax.legend()
ax.grid(alpha=0.2)
ax.set_xlabel("OF score")
ax.set_ylabel("# Pulses")
ax.set_xlim(0,0.4)
```
